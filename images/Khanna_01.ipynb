```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
def true_func(X):
    return np.sin(2 * np.pi * X)

# Chart of fit data
x1 = np.linspace(0,1,10).reshape((-1,1))
y1 = [0.3, 0.9, 1, 0.95, 0, 0.1, -1, -0.5, -0.6, 0.2]
x2 = np.linspace(0,1,100).reshape((-1,1))
y1 = np.array(y1)
```

```python
poly_reg = PolynomialFeatures(degree=0)
x_poly = poly_reg.fit_transform(x1)
poly_reg.fit(x_poly, y1)
lin_reg1 = LinearRegression()
lin_reg1.fit(x_poly, y1)
plt.figure(0)
plt.ylim((-1.5, 1.5))
plt.scatter(x1, y1)
plt.plot(x2, true_func(x2), color="green")
plt.plot(x2,lin_reg1.predict(poly_reg.fit_transform(x2)), color = 'red')
rmse = sqrt(mean_squared_error(x2, lin_reg1.predict(poly_reg.fit_transform(x2))))
print("Root Mean Square Error for Degree '0':", rmse)
train_err0 = mean_squared_error(x2, true_func(x2))
print("Train error for degree '0':", train_err0)
test_err0 = mean_squared_error(x2, lin_reg1.predict(poly_reg.fit_transform(x2)))
print("Test error for degree '0':", test_err0)
```
